```python
%matplotlib inline
import torch
import d2l
```

#### 4.1.1 ReLU函数

```python
x = torch.arange(-8, 8, 0.1, requires_grad=True)
y = torch.relu(x)
```

```python
d2l.plot(x.detach(), y.detach(), 'x', 'relu(x)', figsize=(8, 3.5))
```

```python
# ReLU函数的导数
y.backward(torch.ones_like(x), retain_graph=True)
```

```python
d2l.plot(x.detach(), x.grad, 'x', 'x的导数', figsize=(7, 3))
```

#### 4.1.2 sigmoid函数

```python
x.grad.zero_()
y = torch.sigmoid(x)
```

```python
d2l.plot(x.detach(), y.detach(), 'x', 'sigmoid(x)', figsize=(9, 3))
```

```python
y.backward(torch.ones_like(x), retain_graph=True)
```

```python
d2l.plot(x.detach(), x.grad, 'x', 'sigmoid函数的导数', figsize=(9, 3))
```

#### 4.1.3 tanh函数

```python
x.grad.zero_()
```

```python
y = torch.tanh(x)
```

```python
d2l.plot(x.detach(), y.detach(), 'x', 'tanh(x)', figsize=(9, 4.5))
```

```python
x.grad.data.zero_()
```

```python
y.backward(torch.ones_like(x), retain_graph=True)
```

```python
d2l.plot(x.detach(), x.grad, 'x', 'tanh的导数', figsize=(10, 4.5))
```
