### 3.1.2. 矢量化加速

```python
n = 10000
```

```python
a = torch.ones(n)
b = torch.ones(n)
```

```python
a, b
```

```python
c = torch.zeros(n)
timer = d2l.Timer()
for i in range(n):
    c[i] = a[i] + b[i]
print(f'{timer.stop():0.5f}秒')
```

```python
timer.start()
d = a + b
print(f'{timer.stop():0.5f}秒')
```

```python
c == d
```

### 3.1.3 正态分布和平方损失

```python
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)
```

```python
x = np.arange(-7, 7, 0.01)
```

```python
params = [(0, 1), (0, 2), (3, 1)]
```

```python
y = [normal(x, mu, sigma) for mu, sigma in params]  
```

```python
d2l.plot(x, y, xlabel='x', ylabel='正态分布概率密度函数p(x)', legend=[f'均值={m}，标准差={s}' for m, s in params], figsize=(12, 5))
```
