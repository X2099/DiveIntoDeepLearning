```python
%matplotlib inline
import numpy as np
```

```python
def f(x):
    return 3 * x ** 2 - 4 * x
```

```python
def numerical_lim(f, x, h):
    return (f(x + h) - f(x)) / h
```

```python
h = 0.1
for i in range(5):
    print(f'{h:0.6f} ~ {numerical_lim(f, 1, h):0.6f}')
    h *= 0.1
```

```python
plot(x, [f(x), 2 * x - 3], figsize=(7, 5), legend=['f(x)', '切线（x=1）'])
```
